### Description:

You can get the metainfo "SIFG_info.xlsx" by runing the following code. 

"SIFG_info.xlsx" is used for compute metrics, this can avoid loading npz files during metric calculation.

```python
import pandas as pd
import numpy as np
import glob
```

```python
def read_metadata(path):
    data = np.load(path)
    return list(data['seq_imgs'].shape), list(data["times"]), list(data["labels"])
```

```python
all_files = glob.glob("./*/*.npz")
all_files[0]
```

```python
df = pd.DataFrame(columns=["type","patient_id", "seq_imgs_shape", "times", "labels"])
for idx, each_path in enumerate(all_files):
    info = each_path.split("/")
    data_type, patient_id = info[1], info[2][:-4]
    seq_imgs_shape, times, labels = read_metadata(each_path)#
    df.loc[idx] = [data_type, patient_id, seq_imgs_shape, times, labels]
df
```

```python
df.to_excel("SIGF_info.xlsx",index=False)
```
